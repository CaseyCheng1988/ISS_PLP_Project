```python
import pickle as pk
vectorizer = pk.load(open('vectorizer_shopee_aspect.pk', 'rb'))
```

```python
def logreg_model(TEST_REVIEW):
    loaded_model = pk.load(open('logreg_model_shopee_aspect.sav', 'rb'))
    test_vectors = vectorizer.transform([TEST_REVIEW])
    predME = loaded_model.predict(test_vectors)
    pred= list(predME)
    if pred==[1.0]:
        prediction = "Positive"
    else:
        prediction = "Negative"
    print('Predicted sentiment:%s'%prediction
        )
    if prediction == "Positive":
        return 1
    else:
        return -1
```

```python
TEST_REVIEW = 'Size is not correct'
logreg_model(TEST_REVIEW)
```

```python
def SVM_model(TEST_REVIEW):
    loaded_model = pk.load(open('SVM_model_shopee_aspect.sav', 'rb'))
    test_vectors = vectorizer.transform([TEST_REVIEW])
    predME = loaded_model.predict(test_vectors)
    pred= list(predME)
    if pred==[1.0]:
        prediction = "Positive"
    else:
        prediction = "Negative"
    print('Predicted sentiment:%s'%prediction
        )
    if prediction == "Positive":
        return 1
    else:
        return -1
```

```python
TEST_REVIEW = 'Size is not correct'
SVM_model(TEST_REVIEW)
```

```python
# testing it with amazon_full datasets
import numpy as np
import pandas as pd
with open(r'C:\Users\user\OneDrive - National University of Singapore\EBA5004\project\AMAZON_FASHION.npy', 'rb') as f:
    a = np.load(f, allow_pickle=True)
df_data=pd.DataFrame(a,columns=['asin', 'overall', 'reviewText', 'reviewTime', 'reviewerID',
       'reviewerName', 'summary', 'unixReviewTime', 'verified', 'vote',
       'style', 'image'
])
subset_data=df_data
```

```python
subset_data.loc[subset_data['overall'] <= 3, 'equal_or_lower_than_3?'] = -1
subset_data.loc[subset_data['overall'] > 3, 'equal_or_lower_than_3?'] = 1 
```

```python
X_full_amazon=subset_data['reviewText'].apply(str).tolist()[:1358]
Y_full_amazon=subset_data['equal_or_lower_than_3?'].tolist()[:1358]
len(Y_full_amazon)
```

```python
X_full_amazon[:5]
```

```python
pred_full_amazon_logreg_model = [logreg_model(x_full_amazon) for x_full_amazon in X_full_amazon]
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(Y_full_amazon, pred_full_amazon_logreg_model))
print(metrics.classification_report(Y_full_amazon, pred_full_amazon_logreg_model))
```

```python
pred_full_amazon_svm = [SVM_model(x_full_amazon) for x_full_amazon in X_full_amazon]
```

```python
print(metrics.confusion_matrix(Y_full_amazon, pred_full_amazon_svm))
print(metrics.classification_report(Y_full_amazon, pred_full_amazon_svm))
```
