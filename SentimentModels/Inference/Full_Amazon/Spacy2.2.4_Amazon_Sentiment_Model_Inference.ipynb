```python
import spacy
print(spacy.__version__)
# Load saved model
loaded_model = spacy.load("model_artifacts_full_amazon") 
```

```python
# Test predictions for reviews
def spacy2_model(TEST_REVIEW):
    # Generate prediction
    input_data=TEST_REVIEW
    parsed_text = loaded_model(input_data)
    # Determine prediction to return
    if parsed_text.cats["pos"] > parsed_text.cats["neg"]:
        prediction = "Positive"
        score = parsed_text.cats["pos"]
        overall =score*3+3
    else:
        prediction = "Negative"
        score = parsed_text.cats["neg"]
        overall =3-score*3
    print(f"Review text: {input_data}\nPredicted sentiment: {prediction}"
        f"\tScore: {score}")
    
    print('overall:%s'%overall)
    if prediction == "Positive":
        return 1
    else:
        return -1
```

```python
TEST_REVIEW = "this shirt is not bad"
spacy2_model(TEST_REVIEW)
```

```python
# testing it with shopee_aspect datasets
import json
path=r'C:\Users\user\Documents\GitHub\ISS_PLP_Project\SentimentModels\Training\Shopee_Aspect\absa_textcat.jsonl'
lines = []
with open(path, 'r') as f:
    for line in f:
        lines.append(json.loads(line))
Y_shopee_aspect=[]
X_shopee_aspect=[]
for l in lines:
    X_shopee_aspect.append(l['text'])
    if l['accept']==['POSITIVE']:
        Y_shopee_aspect.append(1)
    else:
        Y_shopee_aspect.append(-1)
```

```python
pred_shopee_aspect = [spacy2_model(x_shopee_aspect) for x_shopee_aspect in X_shopee_aspect]
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(Y_shopee_aspect, pred_shopee_aspect))
print(metrics.classification_report(Y_shopee_aspect, pred_shopee_aspect))
```
