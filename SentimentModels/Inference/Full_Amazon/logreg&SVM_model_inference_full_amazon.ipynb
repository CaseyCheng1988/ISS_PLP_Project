```python
import pickle as pk
vectorizer = pk.load(open('vectorizer_full_amazon.pk', 'rb'))
```

```python
def logreg_model(TEST_REVIEW):
    loaded_model = pk.load(open('logreg_model_full_amazon.sav', 'rb'))
    test_vectors = vectorizer.transform([TEST_REVIEW])
    predME = loaded_model.predict(test_vectors)
    pred= list(predME)
    if pred==[1.0]:
        prediction = "Positive"
    else:
        prediction = "Negative"
    print('Predicted sentiment:%s'%prediction
        )
    if prediction == "Positive":
        return 1
    else:
        return -1
```

```python
TEST_REVIEW = "this shirt is good"
logreg_model(TEST_REVIEW)
```

```python
def SVM_model(TEST_REVIEW):
    loaded_model = pk.load(open('SVM_model_full_amazon.sav', 'rb'))
    test_vectors = vectorizer.transform([TEST_REVIEW])
    predME = loaded_model.predict(test_vectors)
    pred= list(predME)
    if pred==[1.0]:
        prediction = "Positive"
    else:
        prediction = "Negative"
    print('Predicted sentiment:%s'%prediction
        )
    if prediction == "Positive":
        return 1
    else:
        return -1
```

```python
TEST_REVIEW = "this shirt is good"
SVM_model(TEST_REVIEW)
```

```python
# testing it with shopee_aspect datasets
import json
path=r'C:\Users\user\Documents\GitHub\ISS_PLP_Project\SentimentModels\Training\Shopee_Aspect\absa_textcat.jsonl'
lines = []
with open(path, 'r') as f:
    for line in f:
        lines.append(json.loads(line))
Y_shopee_aspect=[]
X_shopee_aspect=[]
for l in lines:
    X_shopee_aspect.append(l['text'])
    if l['accept']==['POSITIVE']:
        Y_shopee_aspect.append(1)
    else:
        Y_shopee_aspect.append(-1)
```

```python
pred_shopee_aspect_logreg = [logreg_model(x_shopee_aspect) for x_shopee_aspect in X_shopee_aspect]
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(Y_shopee_aspect, pred_shopee_aspect_logreg))
print(metrics.classification_report(Y_shopee_aspect, pred_shopee_aspect_logreg))
```

```python
pred_shopee_aspect_svm = [SVM_model(x_shopee_aspect) for x_shopee_aspect in X_shopee_aspect]
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(Y_shopee_aspect, pred_shopee_aspect_svm))
print(metrics.classification_report(Y_shopee_aspect, pred_shopee_aspect_svm))
```
